```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import csv

x=np.arange(0, 5, 0.1)
y=np.sin(x)
a=np.cos(x)
b=np.tan(x)
c=x
```

```python
plt.subplot(411)
plt.plot(y)
plt.subplot(412)
plt.plot(a)
plt.subplot(413)
plt.plot(b)
plt.subplot(414)
plt.plot(c)
```

```python
plt.hold(True)
plt.plot(x,y,'g',linewidth=3)
plt.plot(x,a,'purple',linewidth=1)
plt.plot(x,b,'red',linewidth=1)
plt.plot(x,x,'orange',linewidth=4)
plt.show()
```

```python
with open("Downloads\Temperature.csv",'r') as f:
    reader = csv.reader(f)
    columns = next(reader)
    colmap = dict(zip(columns, range(len(columns))))

arr = np.matrix(np.loadtxt("Downloads\Temperature.csv", delimiter=",", skiprows=1))
Date=(arr[:, colmap['Date']])
Temp=(arr[:, colmap['Temp']])
```

```python
plt.plot(Date, Temp,'green')
plt.xlabel('Date')
plt.ylabel('Temperature')
plt.title('Date vs. Temperature')
plt.show()
```
